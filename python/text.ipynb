```python
import requests
from bs4 import BeautifulSoup

# URL 설정
url = "https://www.zipup.co.kr/"

# 요청 보내기
response = requests.get(url)
response.raise_for_status()  # 요청이 성공하지 않을 경우 예외 발생

# BeautifulSoup으로 HTML 파싱
soup = BeautifulSoup(response.content, "html.parser")

# 페이지에서 모든 텍스트 추출 (태그에 관계없이)
all_text = soup.get_text()

# 특정 태그의 텍스트만 추출 (예: <p> 태그만)
paragraphs = soup.find_all("p")
paragraph_texts = [p.get_text() for p in paragraphs]

# 원하는 텍스트 출력
print(all_text)  # 전체 텍스트
print(paragraph_texts)  # <p> 태그의 텍스트
```

```python
print(paragraph_texts)  # <p> 태그의 텍스트
```

```python
with open("output.txt", mode="w", encoding="utf-8") as file:
    for paragraph in paragraph_texts:
        file.write(paragraph + "\n")
```

```python
from itertools import combinations

def find_combinations(categories_points, target_score):
    all_combinations = []
    categories = list(categories_points.keys())
    
    for r in range(1, len(categories) + 1):
        for combo in combinations(categories, r):
            if sum(categories_points[cat] for cat in combo) == target_score:
                all_combinations.append(combo)
    
    return all_combinations

def main():
    categories_points = {
        "category1": 10,
        "category2": 5,
        "category3": 30,
        "category4": 15,
        "category5": 20,
        "category6": 10,
        "category7": 25,
        "category8": 10,
        "category9": 5,
        "category10": 10,
        "category11": 20
    }
    target_score = 100
    
    all_combinations = find_combinations(categories_points, target_score)
    
    print(f"총 {len(all_combinations)}개의 조합이 발견되었습니다.")
    for combination in all_combinations:
        category_combination = {cat: categories_points[cat] for cat in combination}
        print(category_combination)

# 실행
main()
```
